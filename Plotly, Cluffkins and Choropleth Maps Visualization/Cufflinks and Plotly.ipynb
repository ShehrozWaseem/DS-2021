```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
# agar kch aggregate calc ka kaam ha box plot wu khd krlega 
# or dist draw krke dega uske accordingly
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
# we'll use surface plot to make a 3d vis of this df
df3.iplot(kind='surface')
# u can also change the colorscale of surface plotly by passing colorscale as an arg
```

```python
df.iplot(kind='hist',bins=25)
```

```python
# spread vis for stock data
df[['A','B']].iplot(kind='spread')
# we can observe the spread b/w 2 stocks
```

```python
df.iplot(kind='bubble',x='A',y="B",size="C")
# this is also a type of scatter plot but yhan marker ka size 3rd col ki
# basisi pe define ha ju mene code ma C rkha ha
```
