```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,
                 random_state=101)
# data.shape() krke dekhu 200,2 mileyga yaani 200 features hn hamare pas 

# abi data tuple form ma ha

# data ban ke aya tuple ma ab agar ma data run krnga mje tuple ma milega 
# or ma jesy hi data[0] krnga tu mje array/numpy ma milega
# this numpy array is my no of samples and with 2 col as features
# data[0].shape ab iss data ma mery pas 4 blobs hn 
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
# [:,0] all the rows in first col and plot thos against:
#[:,1] all the rows in 2nd col
#agar ma srf *data[0][:,0],data[0][:,1]* isku plot krdn mje 2 blob milenge
#now the 2nd items in tuple data[1] tu usme 0 1 2 3 huga tu their are 4 centers in the blob
# so i'll do c=data[1] or ab mery pas 4 blob ajaenge
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
# here i know about n_cluster
```

```python
kmeans.fit(data[0])
# features are data[0] that we just plotted earlier
```

```python
kmeans.cluster_centers_
# these are the centers of the 4 clusters
```

```python
kmeans.labels_
# this returns the labels to be true for the clusters

# now if u are working with real data and u wouldn't have the labels so you would have been done
# to compare them to the actual values but since we know the label from data[1] these are *predicted labels kmeans.labels_*
# and now k menas clusterng is unsup so we will not predict the labels but we will try to find the labels in the data
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
# sharey true allow them to hsare same axis

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```
