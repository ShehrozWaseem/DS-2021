```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
axe.set_xlabel('x axis')
axe.set_ylabel('y axis')
axe.set_title('simple graph')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y,'-r')
ax2.plot(y,x,'-r')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('X axis')
ax1.set_ylabel('Z axis')

ax2.plot(x,y)
ax2.set_title('zoom title')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
# as we're using subplot we can itterate
axes[0].plot(x,y,'-b',linewidth=3,ls=':')
axes[1].plot(x,z,'-r',lw=2,ls='--')
```
