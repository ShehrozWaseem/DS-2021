```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# we use this for showing the graph in jupyter notebook agar kisi dusre ide ma kre so use 
# plt.show()
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# functional and oop
plt.plot(x,y,'r-')
# r- color derhe issi trh line style b de skte
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiplot on a same canvas
plt.subplot(1,2,1)
# (no of rows, no of cols, kis no ka graph ha)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'-b')
```

```python
# oo ma hame fig banaenege jiss hame obj miljae consider this obj as an empty canvas 
# ab is canvas pe ham axes ko define krenge
# isme functional wale se kaam zyada ha likn iss plotting method ma hamare pas control huta ha
fig = plt.figure()
axes=fig.add_axes([0.,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LAbel')
axes.set_ylabel('Y LAbel')
axes.set_title('Title')
```

```python
# multiplotting using oop
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
# subplot ma fig or axes ek saat utha ke bana deta
# axes mery pas array of obj huta tu ma usku iterate,indexing kr skta
fig,axes = plt.subplots(nrows=1,ncols=2)
# yani 1 row 2 col 
axes[0].plot(x,y)
axes[1].plot(y,x)
# this is used to avoid overlapping of graphs
plt.tight_layout()
```

```python
axes
# u can see array of abj
```

```python
# figsize and dpi

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square')
ax.plot(x,x**3,label='cube')
# loc ki val docs se dekhu
ax.legend(loc=0)
```
