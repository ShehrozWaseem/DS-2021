```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# means male ke lye bill zyada huga
# female ke lye kam huga
# estimator se std pata lagyi per col
```

```python
sns.countplot(x='sex',data=tips)
# y axis is already choosend and iti will count the occurences
```

```python
# boxplot quantative data ke lye use krte
sns.boxplot(x='day',y='total_bill',data=tips)
# this bp shows the quartile of the dataset
# dots outliers hn mery pas
# hue='smoker' add krdn 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)

# ab isme dekhu jitter ni huga tu scatter plots bht densed hunge or obs shi ni hu paegai
# jitter true krke den inc krte achi obs ke lye
# baki hue add krke dodge true krdya take properly obs kr sken
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
# kind ma violin swormplot ju b user krna bata du
```
